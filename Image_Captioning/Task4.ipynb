```python
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
import glob
import json
import pandas as pd
```

```python
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-large")
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-large")
```

```python
def process(image_path):
    i_image = Image.open(image_path)
    if i_image.mode != "RGB":
        i_image = i_image.convert(mode="RGB")
    inputs = processor(i_image, return_tensors="pt")
    out = model.generate(**inputs)
    res = processor.decode(out[0], skip_special_tokens=True)
    return res
```

```python
from transformers import AutoTokenizer, AutoModel
import torch
import torch.nn.functional as F
from sklearn.metrics.pairwise import cosine_similarity

def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
model = AutoModel.from_pretrained('sentence-transformers/all-MiniLM-L6-v2')
# sentences = ['This is an example sentence', 'Each sentence is converted']



# encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

# with torch.no_grad():
#     model_output = model(**encoded_input)

# sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

# sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)

# print("Sentence embeddings:")
# print(sentence_embeddings.shape)

# similarity_score = cosine_similarity([sentence_embeddings[0]], [sentence_embeddings[1]])[0][0]

# print("Similarity Score:", similarity_score)
```

```python
for i in hemang:
  sentences = [i['text'],i['caption']]
  encoded_input = tokenizer(sentences, padding=True, truncation=True, return_tensors='pt')

  with torch.no_grad():
      model_output = model(**encoded_input)

  sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])
  sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
  i['similarity_score'] = cosine_similarity([sentence_embeddings[0]], [sentence_embeddings[1]])[0][0]
  print(i['similarity_score'])
```
